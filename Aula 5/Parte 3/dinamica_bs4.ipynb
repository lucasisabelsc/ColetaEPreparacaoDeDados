```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
urls = [f'https://www.scrapethissite.com/pages/forms/?page_num={i}' for i in range(1, 6)]
urls
```

```python
r = requests.get('https://www.scrapethissite.com/pages/forms/?page_num=1')
```

```python
soup = BeautifulSoup(r.text, 'html5lib')
```

```python
team = soup.select_one(r'.team')
nome = team.select_one(r'.name').text.strip()
ano = team.select_one(r'.year').text.strip()
```

```python
teams = []
for url in urls:
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'html5lib')
    for team in soup.select('.team'):
        name = team.select_one('.name').text.strip()
        year = team.select_one('.year').text.strip()
        wins = int(team.select_one('.wins').text.strip())
        losses = int(team.select_one('.losses').text.strip())
        ot_losses = team.select_one('.ot-losses').text.strip()
        pct = float(team.select_one('.pct').text.strip())
        gf = int(team.select_one('.gf').text.strip())
        ga = int(team.select_one('.ga').text.strip())
        diff = int(team.select_one('.diff').text.strip())
        teams.append({'name': name,
                    'year': year,
                    'wins': wins,
                    'losses': losses,
                    'ot_losses': ot_losses,
                    'pct': pct,
                    'gf': gf,
                    'ga': ga,
                    'diff': diff})
teams
```

```python
teams[:2]
```

```python
df = pd.DataFrame(teams)
df.head()
```

# Questão 4
Qual o total de gols em cada uma das temporadas

```python
(
    df
    .groupby('year')
    .agg({'gf': 'sum', 'ga': 'sum'})
)
```

# Questão 5
Qual time obteve o maior número de vitórias?

```python
(
    df
    .drop(columns=['year', 'losses', 'ot_losses', 'pct', 'gf', 'ga', 'diff'])
    .groupby('name')
    .sum()
    .sort_values('wins', ascending=False)
)
```

# Questão 6
Qual time obteve o maior número de derrotas?

```python
(
    df
    .drop(columns=['year', 'wins', 'ot_losses', 'pct', 'gf', 'ga', 'diff'])
    .groupby('name')
    .sum()
    .sort_values('losses', ascending=False)
)
```
