```python
import pandas as pd
```

```python
df = pd.read_csv('./emails.csv', sep=',')
df.head()
```

Podemos usar o método extract para extrair o nome do email, o provedor e o *top level domain* da coluna email usando uma expressão regular.

```python
(
    df
    .email
    .str
    .extract(r'([\w\.]+)@([\w]+)\.(\w+)')
    .rename(columns={0: 'user', 1: 'domain', 2: 'ext'})
)
```
