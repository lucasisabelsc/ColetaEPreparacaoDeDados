```python
import requests
import xml.etree.ElementTree as ET
```

```python
r = requests.get('''http://g1.globo.com/sitemap/g1/2022/12/09_1.xml''')
print(r.status_code)
print(r.headers)
```

```python
root = ET.fromstring(r.text, parser=ET.XMLParser(encoding='utf-8'))
for child in root.iter('{http://www.sitemaps.org/schemas/sitemap/0.9}loc'):
    print(child.text)
```
