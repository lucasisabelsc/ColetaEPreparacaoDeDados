```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
salario = pd.Series([260, 300, 320, 320, 380, 400, 400, 420, 420, 440, 440, 500,
                          500, 520, 600, 660, 660, 680, 700, 700, 740, 740, 740, 900, 
                          920, 1040, 1400])
```

```python
salario.describe().round(2)
```

```python
salario.var(ddof=1).round(2)
```

```python
salario.std(ddof=1).round(2)
```

```python
sns.histplot(salario, bins=10)
plt.show()
```

```python
sns.boxplot(salario, orient='h')
plt.show()
```
