```python
import pandas as pd
import requests
import json
pd.set_option('display.max_columns', None)
```

Requisição para API

```python
r = requests.get(r'http://127.0.0.1:5001/country')
r = r.json()
df_01 = pd.DataFrame.from_dict(r, orient='index')
df_01.reset_index(inplace=True)
df_01.rename(columns={'index': 'country_code'}, inplace=True)
df_01
```

Leitura de arquivo CSV

```python
df_02 = pd.read_csv('./fase01/gdp_data.csv')
df_02
```

Merge dos DataFrames pela coluna Country Code

```python
df = pd.merge(df_01, df_02, left_on='country_code', right_on='Country Code', how='inner')
df
```

Organização e limpeza do DF

```python
df_limpo = df.drop(columns=['Country Code', 'Country Name', 'Indicator Name', 'Indicator Code', 'currency', 'Unnamed: 66'])
df_limpo.info()
df_limpo
```

Exportação CSV

```python
df_limpo.to_csv('./data_projeto.csv', index=False)
```
